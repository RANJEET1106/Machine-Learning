```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('Datasets/df1',index_col=0)
df1.head()
```

```python
df2 = pd.read_csv('Datasets/df2')
df2.head()
```

```python
plt.figure()
df1['A'].hist(bins=30)
plt.show()
plt.close()
```

```python
plt.figure()
df1['A'].plot(kind='hist',bins=30)
plt.show()
plt.close()
```

```python
plt.figure()
df1['A'].plot.hist()
plt.show()
plt.close()
```

```python
plt.figure()
df2.plot.area(alpha=0.4)
plt.show()
plt.close()
```
