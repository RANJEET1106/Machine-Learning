```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

# Distribution Plot

```python
plt.figure()
sns.histplot(tips['total_bill'],kde=True,bins=30)  # distplot is deprecated
plt.show()
plt.close()
```

```python
plt.figure()
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
plt.close()
```

```python
plt.figure()
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
plt.show()
plt.close()
```

```python
plt.figure()
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
plt.show()
plt.close()
```

```python
plt.figure()
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
plt.show()
plt.close()
```

```python
plt.figure()
sns.pairplot(tips)
plt.show()
plt.close()
```

```python
plt.figure()
sns.pairplot(tips,hue='sex',palette='coolwarm') 
# hue can take categorical arguments
plt.show()
plt.close()
```

```python
plt.figure()
sns.rugplot(tips['total_bill'])
plt.show()
plt.close()
```

```python
plt.figure()
sns.kdeplot(tips['total_bill'])
plt.show()
plt.close()
```

# Matrix Plot

```python
tc =tips.corr(numeric_only=True)
```

```python
plt.figure()
sns.heatmap(tc,annot=True,cmap='coolwarm')
plt.show()
plt.close()
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers',observed=False)
```

```python
plt.figure()
sns.heatmap(fp,cmap='coolwarm',linecolor='white',linewidths=2)
plt.show()
plt.close()
```

```python
plt.figure()
sns.clustermap(fp,cmap='coolwarm')
plt.show()
plt.close()
```

```python
plt.figure()
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
plt.show()
plt.close()
```

# Grids Plot

```python
plt.figure()
sns.pairplot(iris)
plt.show()
plt.close()
```

```python
plt.figure()
g= sns.PairGrid(iris)
g.map_diag(sns.histplot, kde=True)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
plt.show()
plt.close()
```

```python
plt.figure()
g= sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.histplot,'total_bill',kde=True)
plt.show()
plt.close()
```

```python
plt.figure()
g= sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
plt.show()
plt.close()
```

# Regression Plot

```python
plt.figure()
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v']
          ,scatter_kws={'s':100})
plt.show()
plt.close()
```

```python
plt.figure()
sns.lmplot(x='total_bill',y='tip',data=tips,col='time',row='day',hue='sex'
            ,aspect=0.6,height=5)
plt.show()
plt.close()
```

# Style and Color
